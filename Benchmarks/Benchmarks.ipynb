```python
import matplotlib.pyplot as plt
import numpy as np
```

# CPU Usage Plot

```python
with open("Forward_Euler_Bench/Forward_Euler_CPU_Usage.txt", "r") as FECPUs:
    fecpu_x, fecpu_y = FECPUs.read().splitlines()
    fecpu_x =  fecpu_x.split("\t")
    fecpu_y =  fecpu_y.split("\t")
    fecpu_x, fecpu_y = [float(h) for h in fecpu_x], [float(z) for z in fecpu_y]
    
    
with open("Rungekutta/RUNGEKUTTA_CPU_Usage.txt","r") as RKCPU:
    rkcpu_x, rkcpu_y = RKCPU.read().splitlines()
    rkcpu_x =  rkcpu_x.split("\t")
    rkcpu_y =  rkcpu_y.split("\t")
    rkcpu_x, rkcpu_y = [float(hh) for hh in rkcpu_x], [float(zz) for zz in rkcpu_y]

with open("simtime_50_NMR/Next_Reaction_Method_CPU_Usage.txt","r") as NRMCPU:
    nrmcpu_x, nrmcpu_y = NRMCPU.read().splitlines()
    nrmcpu_x = nrmcpu_x.split("\t")
    nrmcpu_y =  nrmcpu_y.split("\t")
    nrmcpu_x, nrmcpu_y = [float(hhh) for hhh in nrmcpu_x], [float(zzz) for zzz in nrmcpu_y]
    
    
```

```python
plt.figure(figsize=(10,10))


names = ["FEuler", "NRM", "RK4"]


values_cpu = [np.median(fecpu_x), np.median(nrmcpu_x), np.median(rkcpu_x)]

plt.bar(names, values_cpu, color=("red", "gray", "green"))
plt.ylim((97,98))
plt.xticks(fontsize=20)
# plt.plot(nrmcpu_x, nrmcpu_y, label='Next Reaction Method CPU Usage in %')
# plt.plot(rkcpu_x, rkcpu_y, label='Runge Kutta CPU Usage in %')
# plt.legend()

plt.show()
```

# Time of Execution 

```python
times = [5.1836206912994385, 2.114501953125, 20.650723934173584]
plt.figure(figsize=(10,10))
plt.bar(names, times, color=("red", "gray", "green"))
plt.text(-0.5, 21, 'NRM simulation time = 50')
plt.xticks(fontsize=20)
plt.show()
```

# Memory Usage

```python
with open("Forward_Euler_Bench/Forward_Euler_Memory.txt", "r") as FEMEMs:
    femem_x, femem_y = FEMEMs.read().splitlines()
    femem_x =  femem_x.split("\t")
    femem_y =  femem_y.split("\t")
    femem_x, femem_y = [float(h2) for h2 in femem_x], [float(z2) for z2 in femem_y]
    
    
with open("Rungekutta/RUNGEKUTTA_Memory.txt","r") as RKMEM:
    rkmem_x, rkmem_y = RKMEM.read().splitlines()
    rkmem_x =  rkmem_x.split("\t")
    rkmem_y =  rkmem_y.split("\t")
    rkmem_x, rkmem_y = [float(hh2) for hh2 in rkmem_x], [float(zz2) for zz2 in rkmem_y]

with open("simtime_50_NMR/Next_Reaction_Method_Memory.txt","r") as NRMMEM:
    nrmmem_x, nrmmem_y = NRMMEM.read().splitlines()
    nrmmem_x = nrmmem_x.split("\t")
    nrmmem_y =  nrmmem_y.split("\t")
    nrmmem_x, nrmmem_y = [float(hhh2) for hhh2 in nrmmem_x], [float(zzz2) for zzz2 in nrmmem_y]
```

```python
plt.figure(figsize=(10,10))


names = ["FEuler", "NRM", "RK4"]


values = [np.median(femem_x), np.median(nrmmem_x), np.median(rkmem_x)]

plt.bar(names, values, color=("red", "gray", "green"))
#plt.ylim((97,98))
plt.xticks(fontsize=20)
# plt.plot(nrmcpu_x, nrmcpu_y, label='Next Reaction Method CPU Usage in %')
# plt.plot(rkcpu_x, rkcpu_y, label='Runge Kutta CPU Usage in %')
# plt.legend()

plt.show()
```
